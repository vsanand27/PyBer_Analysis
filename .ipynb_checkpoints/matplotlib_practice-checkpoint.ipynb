```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis,y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Add a grid
plt.grid()
```

```python
# Using matlab  approach "object Oriented"
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis,marker="*", color="blue",label='Boston')
#Set y axis 0 to 45
ax.set_ylim(0, 45)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

Scatter Plot Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis, 'o')
#Note "o" changes from line graph to scatter plot graph
```

Ask about Inversion of y_axis

```python
plt.scatter(y_axis,x_axis, color="green", s=y_axis, label='Chicago')
plt.gca().invert_yaxis() 
plt.legend()
plt.xlabel('Fare ($)')
plt.ylabel('Month')
plt.title('PyBer Fare by Month')
plt.xlim(0,50)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
plt.scatter(y_axis,x_axis, color="green", s=y_axis_larger, label='Chicago')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel('Fare ($)')
plt.ylabel('Month')
plt.title('PyBer Fare by Month')
plt.xlim(0,50)
```

```python
#new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots(figsize=(8,7))
ax.scatter(y_axis,x_axis, c='skyblue', s= [i*5 for i in y_axis],alpha=.2,edgecolors='black', linewidths=5,label="Boston")
ax.legend()
# ask if this okay to invert for alternative method
plt.gca().invert_yaxis()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Month')
ax.set_title('PyBer Fare by Month')

```

# Pie Charts

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.5, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,labels=x_axis,autopct='%.1f%%',explode=explode_values)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Change the size of the figure so that % don't run over each other
plt.subplots(figsize=(8,8))

plt.pie(y_axis,labels=x_axis,autopct='%.1f%%',explode=explode_values, colors=colors)
plt.show()


```

# Object oriented Pie Chart

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis,autopct='%.1f%%',explode=explode_values,shadow=True,startangle=90,counterclock=False,colors=colors)
plt.show()
```
